## 3. Operaciones matemáticas

Python funciona como una calculadora. Pero mucho más poderoso.

```python
# Operaciones básicas
suma        = 10 + 5      # 15
resta       = 10 - 5      # 5
producto    = 10 * 5      # 50
division    = 10 / 3      # 3.333...
potencia    = 2 ** 8      # 256 (2 elevado a la 8 → niveles de gris de 8 bits)
modulo      = 10 % 3      # 1 (residuo de la división)

print("Suma:", suma)
print("División:", division)
print("2^8 (niveles en 8 bits):", potencia)

# Cálculo del NDVI como operación matemática
# NDVI = (NIR - Rojo) / (NIR + Rojo)
nir   = 3200.0    # valor típico de reflectancia NIR en vegetación sana (Sentinel-2 L2A)
rojo  = 500.0     # valor típico de reflectancia roja en vegetación sana

ndvi = (nir - rojo) / (nir + rojo)
print("\nNIR =", nir, "| Rojo =", rojo)
print("NDVI calculado:", round(ndvi, 4))   # round() redondea a 4 decimales
```

## 4. Listas — colecciones ordenadas de valores

Una lista guarda varios valores en orden. Se define con corchetes `[ ]`.

```python
# Lista de bandas de Sentinel-2
bandas_sentinel2 = ['B1', 'B2', 'B3', 'B4', 'B5', 'B6', 'B7', 'B8', 'B8A', 'B9', 'B11', 'B12']

# Lista de valores NDVI de diferentes coberturas
ndvi_coberturas = [-0.05, 0.05, 0.25, 0.45, 0.65, 0.80]
nombres_cobertura = ['Agua', 'Suelo desnudo', 'Pastura escasa', 'Cultivo joven', 'Cacao adulto', 'Bosque denso']

# Acceder a un elemento por su posición (empieza en 0, no en 1)
print("Primera banda:", bandas_sentinel2[0])     # B1
print("Cuarta banda:",  bandas_sentinel2[3])     # B4 (banda roja)
print("Última banda:",  bandas_sentinel2[-1])    # B12 (SWIR)

# Cuántos elementos tiene la lista
print("\nSentinel-2 tiene", len(bandas_sentinel2), "bandas multiespectrales")
```

```python
# Recorrer una lista con un bucle for
print("Interpretación de valores NDVI:")
print("-" * 45)

for i in range(len(ndvi_coberturas)):
    ndvi_val = ndvi_coberturas[i]
    cobertura = nombres_cobertura[i]
    print(f"  NDVI = {ndvi_val:+.2f}  →  {cobertura}")
```

---

## 5. Condicionales — tomar decisiones

Los condicionales permiten que el código haga cosas distintas según una condición.

```python
# Función para interpretar un valor de NDVI
def interpretar_ndvi(ndvi):
    """Devuelve la interpretación de cobertura para un valor de NDVI."""
    if ndvi < 0:
        return "Agua, nieve o nube"
    elif ndvi < 0.1:
        return "Suelo desnudo, roca o zona urbana"
    elif ndvi < 0.2:
        return "Vegetación muy escasa o muy estresada"
    elif ndvi < 0.4:
        return "Vegetación moderada (pastura, cultivos jóvenes)"
    elif ndvi < 0.6:
        return "Vegetación densa (cultivos en desarrollo)"
    else:
        return "Vegetación muy densa y sana (bosque, cacao adulto)"

# Probar con distintos valores
valores_prueba = [-0.1, 0.05, 0.30, 0.55, 0.75]

print("Interpretación de valores NDVI de la SNSM:")
for v in valores_prueba:
    print(f"  NDVI = {v:.2f}  →  {interpretar_ndvi(v)}")
```

## 6. Importar librerías

Python base es poderoso, pero su verdadero potencial está en las **librerías** —
paquetes de funciones ya escritas por otros que puedes reutilizar.

Las librerías esenciales para teledetección:

| Librería | Para qué sirve |
|----------|----------------|
| `numpy` | Cálculos numéricos rápidos (arrays, matrices) |
| `matplotlib` | Gráficos y visualizaciones |
| `rasterio` | Leer y escribir imágenes satelitales (GeoTIFF, JP2) |
| `geopandas` | Trabajar con shapefiles y datos vectoriales |
| `ee` | Google Earth Engine Python API |
| `geemap` | Mapas interactivos sobre GEE en Colab |

```python
# Importar librerías (ya vienen instaladas en Google Colab)
import numpy as np         # 'as np' es un alias para escribir menos
import matplotlib.pyplot as plt

# Numpy: operaciones sobre arrays (vectores y matrices)
# Imagina un array como una lista pero mucho más potente para matemáticas

# Crear un array de valores NDVI para una franja de píxeles
ndvi_pixeles = np.array([0.12, 0.45, 0.78, 0.65, 0.82, 0.30, -0.05, 0.71, 0.60, 0.88])

print("Array de NDVI:", ndvi_pixeles)
print("Promedio:",      round(np.mean(ndvi_pixeles), 3))
print("Mínimo:",        round(np.min(ndvi_pixeles), 3))
print("Máximo:",        round(np.max(ndvi_pixeles), 3))
print("Desv. estándar:",round(np.std(ndvi_pixeles), 3))

# Operaciones sobre todo el array de una vez (sin bucle for)
ndvi_validos = ndvi_pixeles[ndvi_pixeles > 0]  # filtrar solo valores positivos
print("\nNDVI de píxeles con vegetación (> 0):", ndvi_validos)
```

## 7. Visualización con Matplotlib

Matplotlib es la librería estándar de Python para hacer gráficos.

```python
# Gráfico 1: Firma espectral simplificada de tres coberturas
# (reflectancia relativa en las principales bandas de Sentinel-2)

bandas = ['B2\n(Azul)', 'B3\n(Verde)', 'B4\n(Rojo)', 'B8\n(NIR)', 'B11\n(SWIR)']
posiciones = [0.49, 0.56, 0.665, 0.842, 1.61]  # longitud de onda en µm

# Reflectancias típicas (valores 0-1)
vegetacion_sana  = [0.04, 0.10, 0.05, 0.45, 0.20]
suelo_desnudo    = [0.10, 0.13, 0.15, 0.22, 0.30]
agua_clara       = [0.08, 0.05, 0.02, 0.01, 0.00]

plt.figure(figsize=(10, 5))
plt.plot(posiciones, vegetacion_sana, 'g-o', linewidth=2, markersize=8, label='Vegetación sana (cacao adulto)')
plt.plot(posiciones, suelo_desnudo,   's--', color='sienna', linewidth=2, markersize=8, label='Suelo desnudo')
plt.plot(posiciones, agua_clara,      'b-^', linewidth=2, markersize=8, label='Agua clara')

# Resaltar las bandas usadas en el NDVI
plt.axvspan(0.63, 0.69, alpha=0.15, color='red',   label='Banda Roja (B4)')
plt.axvspan(0.78, 0.90, alpha=0.15, color='green', label='Banda NIR (B8)')

plt.xlabel('Longitud de onda (µm)', fontsize=12)
plt.ylabel('Reflectancia', fontsize=12)
plt.title('Firmas Espectrales — Sierra Nevada de Santa Marta\n(Sentinel-2 bandas seleccionadas)', fontsize=13)
plt.legend(fontsize=10)
plt.xticks(posiciones, bandas)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print("Observa: la vegetación sana tiene BAJA reflexión en Rojo (B4) y ALTA en NIR (B8)")
print("Esa diferencia es exactamente lo que el NDVI captura.")
```

```python
# Gráfico 2: Distribución de valores NDVI (histograma)

# Simulamos valores NDVI de una imagen de la zona cacaotera
np.random.seed(42)

# Mezcla de coberturas: bosque + cacao + zonas de transición + suelo
ndvi_bosque_cacao = np.random.normal(loc=0.72, scale=0.08, size=1500)  # bosque/cacao
ndvi_transicion   = np.random.normal(loc=0.45, scale=0.10, size=800)   # zona de transición
ndvi_suelo        = np.random.normal(loc=0.12, scale=0.06, size=300)   # suelo/urbano
ndvi_agua         = np.random.normal(loc=-0.05, scale=0.04, size=100)  # agua

ndvi_imagen = np.clip(
    np.concatenate([ndvi_bosque_cacao, ndvi_transicion, ndvi_suelo, ndvi_agua]),
    -1, 1
)

plt.figure(figsize=(10, 4))
plt.hist(ndvi_imagen, bins=60, color='forestgreen', edgecolor='white', alpha=0.8)

# Líneas de referencia
plt.axvline(x=0,    color='blue',  linestyle='--', linewidth=1.5, label='NDVI = 0 (límite agua/suelo)')
plt.axvline(x=0.4,  color='orange',linestyle='--', linewidth=1.5, label='NDVI = 0.4 (cultivo moderado)')
plt.axvline(x=0.6,  color='red',   linestyle='--', linewidth=1.5, label='NDVI = 0.6 (vegetación densa)')

plt.xlabel('Valor de NDVI', fontsize=12)
plt.ylabel('Número de píxeles', fontsize=12)
plt.title('Distribución de NDVI — Zona Cacaotera SNSM (simulado)', fontsize=13)
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print(f"Total de píxeles: {len(ndvi_imagen):,}")
print(f"NDVI promedio: {np.mean(ndvi_imagen):.3f}")
print(f"Píxeles con vegetación densa (NDVI > 0.6): {np.sum(ndvi_imagen > 0.6):,} ({100*np.mean(ndvi_imagen > 0.6):.1f}%)")
```

```python
import geopandas as gpd
from shapely.geometry import Polygon

# Create a simple polygon (example)
polygon = Polygon([(0, 0), (1, 1), (1, 0), (0, 0)])

# Create a GeoDataFrame
gdf = gpd.GeoDataFrame({'id': [1]}, geometry=[polygon], crs="EPSG:4326")

# Save the GeoDataFrame to a shapefile
gdf.to_file("parcelas_example.shp", driver='ESRI Shapefile')

print("Archivo parcelas_example.shp creado.")
```

```python
%pip install sentinelsat
from sentinelsat import SentinelAPI, geojson_to_wkt
from datetime import date
import geopandas as gpd

# --- CREDENCIALES DE ACCESO AL PORTAL COPERNICUS ---
# Puedes registrarte gratis en: https://scihub.copernicus.eu/dhus
api = SentinelAPI('f3beff07-9960-4aaa-b307-ed5f104bb705', 'Copernicus12-', 'https://apihub.copernicus.eu/apihub')

# --- LEER SHAPEFILE DE TU ÁREA DE INTERÉS (AOI) ---
gdf = gpd.read_file('parcelas_example.shp')
aoi = geojson_to_wkt(gdf.__geo_interface__)

# --- BÚSQUEDA DE PRODUCTOS SENTINEL-2 ---
products = api.query(
    area=aoi,
    date=('20240101', '20240531'),  # Rango de fechas: enero a mayo de 2024
    platformname='Sentinel-2',
    producttype='S2MSI2A',          # L2A: corrección atmosférica incluida
    cloudcoverpercentage=(0, 20)    # Máximo 20% de nubes
)

# --- IMPRIMIR RESULTADOS ---
print(f"{len(products)} productos encontrados.")
```
